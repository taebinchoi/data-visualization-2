```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
real_gdp = [2900, 2998, 3083, 3260, 3391, 3493, 3532, 3532, 3530, 3656, 3628]

# Real GDP plot
plt.plot(years, real_gdp, label='Real GDP per capita (10,000 KRW)')
plt.xlabel('Year')
plt.ylabel('Real GDP per capita (10,000 KRW)')
plt.title('Trends in Real GDP per capita')
plt.legend()
plt.show()
```
