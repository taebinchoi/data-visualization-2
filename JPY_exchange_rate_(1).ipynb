```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    '년도분기': ['2019.1/4', '2019.2/4', '2019.3/4', '2019.4/4', '2020.1/4', '2020.2/4', '2020.3/4', '2020.4/4',
                '2021.1/4', '2021.2/4', '2021.3/4', '2021.4/4', '2022.1/4', '2022.2/4', '2022.3/4', '2022.4/4', '2023.1/4'],
    '값': [1006, 1092, 1138, 1077, 1109, 1119, 1108, 1049, 1047, 1012, 1053,
          1041, 1047, 968, 973, 967, 954]
}

df = pd.DataFrame(data)

plt.plot(df['년도분기'], df['값'], color='purple', marker='o', linestyle='--')
plt.xlabel('quarter')
plt.ylabel('exchange rate')
plt.title('JPY exchange rate(on a 100-yen basis)')
plt.xticks(rotation=45)
plt.show()
```
