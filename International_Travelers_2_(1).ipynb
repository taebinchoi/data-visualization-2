```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
travelers = [1484.6, 1608.1, 1931.0, 2238.3, 2649.6, 2869.6, 2871.4, 427.6, 122.3, 655.4]

plt.plot(years, travelers, color='orange', label='The number of international travelers (10,000)', marker='o', linestyle='--')
plt.xlabel('Year')
plt.ylabel('Travelers (10,000)')
plt.title('The number of international travelers')
plt.legend()
plt.show()
```
