```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    '연도분기': ['2021.1/4', '2021.2/4', '2021.3/4', '2021.4/4', '2022.1/4', '2022.2/4', '2022.3/4', '2022.4/4', '2023.1/4'],
    'travelers': [5401, 2861, 5679, 5006, 8871, 27672, 81663, 894545, 1600673]
}

df = pd.DataFrame(data)

plt.plot(df['연도분기'], df['travelers'], color='darkgrey', marker='o', linestyle='--')
plt.xlabel('quarter')
plt.ylabel('travelers')
plt.title('Korean tourists visiting Japan')
plt.xticks(rotation=45)
plt.show()
```
