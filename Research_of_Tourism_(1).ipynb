```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
ratio = [60.4, 23.5, 11.7, 3.2, 1.2]
labels = ['China', 'Japan', 'USA', 'Hong Kong', 'Taiwan']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, shadow=True)
plt.show()
```

```python
ratio = [8.5, 77.5, 14.0]
labels = ['(1)', '(2)', '(3)']
colors = ['gold', 'violet', 'beige']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, shadow=True, colors=colors)
plt.show()
```
